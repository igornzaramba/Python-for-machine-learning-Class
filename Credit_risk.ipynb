```python
import pandas as pd

from sklearn.datasets import fetch_openml

dataframe = fetch_openml(data_id=31, target_column=None)

dataframe.frame.info()
```

```python
df=dataframe.data
df
```

```python
from sklearn.preprocessing import LabelEncoder
df["class"]= LabelEncoder().fit_transform(df["class"])
df.head(5)
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=["float64", "int64",'int32'])
sns.heatmap(numeric_df.corr(),annot=True)
```

```python
sns.pairplot(df, hue='class')
```

```python
nominal = df[['credit_amount','age','existing_credits','num_dependents']]
categorical = df[['credit_history','purpose','housing','job']]
```

```python
categorical
```

```python
categorical= pd.get_dummies(categorical).astype(int)
```

astype(int) not to get bolean

```python
categorical
```

```python
credits = pd.concat([nominal,categorical],axis=1)
credits
```

```python
y = df['class'].values

x_columns = len(credits.columns)

x = credits.iloc[:,0:x_columns].values
```

seprate the datasent in train, validate  and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=0)
X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test,test_size=0.5,random_state=0)
```

```python
print(len(X_train),len(X_test),len(X_validate))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
knn.score(X_test,y_test)
```

```python
scores = []
neighbors = range (1,20)
for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    results = knn.score(X_test,y_test)
    scores.append(results)
print(scores)
```

```python
plt.plot(neighbors,scores)
```

```python
knn = KNeighborsClassifier(n_neighbors=13).fit(X_train,y_train)
result = knn.score(X_validate, y_validate)
print('accuracy =',result)
```

```python
from sklearn.metrics import confusion_matrix
predictions = knn.predict(X_validate)
print('confusion Matrix: \n', confusion_matrix(y_validate, predictions))
```
